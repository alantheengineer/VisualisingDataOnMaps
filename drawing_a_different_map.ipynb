```python
import os
import shapefile as shp  # Requires the pyshp package
import matplotlib.pyplot as plt

font = {'family': 'sans-serif',
        'color':  'black',
        'weight': 'normal',
        'size': 24,
        }

mex_states_loc=os.path.join(os.getcwd(),'_data','mexstates')
mex_states_pic=os.path.join(os.getcwd(),'_output','states_map.png')
tequila_states=['Tamaulipas','Jalisco','Michoacan','Guanajuato','Nayarit']
tequila_colours=["#b6d7ba","#debbd4","#a9d5df","#e1c2aa","#acbcdf"]
sf = shp.Reader(mex_states_loc)
ax = plt.figure(figsize=(20,14))
for n,shape in enumerate(sf.shapeRecords()):
    rec=sf.record(n)
    dct=rec.as_dict()
    colour='k'
    if dct['ADMIN_NAME'] not in tequila_states:
#        colour='r'
        for i in range(len(shape.shape.parts)):
            i_start = shape.shape.parts[i]
            if i==len(shape.shape.parts)-1:
                i_end = len(shape.shape.points)
            else:
                i_end = shape.shape.parts[i+1]
            x = [i[0] for i in shape.shape.points[i_start:i_end]]
            y = [i[1] for i in shape.shape.points[i_start:i_end]]
            plt.plot(x,y,color='darkblue',linewidth=0.5)
#loop over the shapes again to plot the ones i want on top
col=0
for n,shape in enumerate(sf.shapeRecords()):
    rec=sf.record(n)
    dct=rec.as_dict()
    colour='r'
    if dct['ADMIN_NAME'] in tequila_states:
        c=tequila_colours[col]
        col=col+1
        for i in range(len(shape.shape.parts)):
            i_start = shape.shape.parts[i]
            if i==len(shape.shape.parts)-1:
                i_end = len(shape.shape.points)
            else:
                i_end = shape.shape.parts[i+1]
            x = [i[0] for i in shape.shape.points[i_start:i_end]]
            y = [i[1] for i in shape.shape.points[i_start:i_end]]
            #plt.plot(x,y,c=colour,linewidth=0.8)
            plt.fill(x,y,color=c,label=dct['ADMIN_NAME'])
plt.xlabel("longtitude",fontdict=font)
plt.ylabel("latitude",fontdict=font)
plt.tick_params(axis='both', which='major', labelsize=18, color='grey')
plt.grid(color='grey', linestyle=':', linewidth=0.2)
#plt.legend(fontsize=18)
handles, labels = plt.gca().get_legend_handles_labels()
by_label = dict(zip(labels, handles))
plt.legend(by_label.values(), by_label.keys(),fontsize=18)
plt.savefig(mex_states_pic,bbox_inches='tight')
```
